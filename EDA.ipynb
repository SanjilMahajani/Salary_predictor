```python
sorted_counts = df1['company_name'].value_counts().sort_values(ascending=False)

# Grouping by company_name and aggregating job_title
grouped = df1.groupby('company_name')['job_title'].unique()

# Reordering based on sorted counts
grouped_sorted = grouped.reindex(sorted_counts.index)

# Compute the length of the list of job titles
job_title_lengths = grouped_sorted.apply(len)

# Create a DataFrame with company_name, job_titles, and length of job titles
result_df = pd.DataFrame({
    'company_name': grouped_sorted.index,
    'job_titles': grouped_sorted.values,
    'num_job_titles': job_title_lengths.values
})

# Print the result DataFrame
print(result_df)
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif 'data analyst' in title.lower():
        return 'Data Analyst'
    elif 'machine learning engineer' in title.lower():
        return 'Machine Learning Engineer'
    elif 'business analyst' in title.lower():
        return 'Business Analyst'
    elif 'data architect' in title.lower():
        return 'Data Architect'
    else:
        return 'na'
```

```python
df['job'] = df['job_title'].apply(title_simplifier)
```

```python
df
```

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'Senior'
    else:
        return 'Junior'
```

```python
df['seniority'] = df['job_title'].apply(seniority)
```
